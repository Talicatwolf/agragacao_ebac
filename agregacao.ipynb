```python
import pandas as pd
df_video = pd.read_parquet('/content/videos-preparados.snappy.parquet')
df_video.head()
```

```python
keyword_counts = df_video['Keyword'].value_counts()
print("Keyword Counts:")
print(keyword_counts)
```

```python
interaction_mean_by_keyword = df_video.groupby('Keyword')['Interaction'].mean()
print("Interaction Mean by Keyword:")
print(interaction_mean_by_keyword)
```

```python
rank_interactions = df_video.groupby('Keyword')['Interaction'].max().rename('Rank Interactions').sort_values(ascending=False)
print("Rank Interactions by Keyword (Max Interaction):")
print(rank_interactions)
```

```python
views_stats_by_keyword = df_video.groupby('Keyword')['Views'].agg(['mean', 'var'])
print("Views Statistics by Keyword (Mean and Variance):")
print(views_stats_by_keyword)
```

```python
views_min_max_mean_by_keyword = df_video.groupby('Keyword')['Views'].agg(['mean', 'min', 'max']).astype(int)
print("Views Statistics by Keyword (Min, Max, Mean - No Decimals):")
print(views_min_max_mean_by_keyword)
```

```python
published_at_range_by_keyword = df_video.groupby('Keyword')['Published At'].agg(['min', 'max'])
print("Published At Range by Keyword:")
print(published_at_range_by_keyword)
```

```python
total_titles = df_video['Title'].count()
unique_titles = df_video['Title'].nunique()

print(f"Total number of titles: {total_titles}")
print(f"Number of unique titles: {unique_titles}")

if total_titles == unique_titles:
    print("All titles are unique.")
else:
    print(f"There are {total_titles - unique_titles} duplicate titles.")
```

```python
df_video['Published At'] = pd.to_datetime(df_video['Published At'])
year_counts = df_video['Published At'].dt.year.value_counts().sort_index(ascending=True)
print("Records by Year (Ascending Order):\n")
print(year_counts)
```

```python
df_video['Published YearMonth'] = df_video['Published At'].dt.strftime('%Y-%m')
year_month_counts = df_video['Published YearMonth'].value_counts().sort_index(ascending=True)
print("Records by Year and Month (Ascending Order):\n")
print(year_month_counts)
```

```python
df_video['Year'] = df_video['Published At'].dt.year

df_temp = df_video.sort_values(by=['Keyword', 'Published At'])
df_temp['Cumulative Likes Mean'] = df_temp.groupby('Keyword')['Likes'].expanding().mean().reset_index(level=0, drop=True)

print("Cumulative Mean of Likes by Keyword Over Years:")
print(df_temp[['Keyword', 'Year', 'Cumulative Likes Mean']])
```
